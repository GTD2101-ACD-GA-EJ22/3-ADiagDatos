## Dividendo el Daset para Datos de Validaci√≥n

Para mejorar el desempe√±o de los modelos es necesario entrenarlo con un conjunto de datos llamados de validaci√≥n

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# Define model
melbourne_model = DecisionTreeRegressor()
# Fit model
melbourne_model.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = melbourne_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

Es posible hacer un benchmarking con diferentes configuraciones de profundidad para los √°rboles de decisi√≥n a fin de determinar aqu√©lla que permita el mejor desempe√±o y que ofrezca tambi√©n el mejor valor entre el underfitting y el overfitting del modelo.

```python
from sklearn.metrics import mean_absolute_error
from sklearn.tree import DecisionTreeRegressor

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
# compare MAE with differing values of max_leaf_nodes
for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

Se puede concluir que el n√∫mero √≥ptimo de hojas en el √°rbol de decisi√≥n es 500 hojas. Donde se observa que el modelo ofrece un error promedio m√≠nimo.

# Random Forest

En lugar de crear modelos con 1 s√≥lo √°rbol de decisi√≥n, podemos crear un modelo que sea un pron√≥stico basado en el promedio de m√∫ltiples √°rboles de decisi√≥n. Eso es lo que hace el modelo de Random Forest.

```python
import pandas as pd
    
# Load data
melbourne_file_path = '/content/drive/MyDrive/Colab Notebooks/APD/datasets/melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path) 

# Filter rows with missing values
melbourne_data = melbourne_data.dropna(axis=0)

# Choose target and features
y = melbourne_data.Price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 
                        'YearBuilt', 'Lattitude', 'Longtitude']
X = melbourne_data[melbourne_features]

from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor(random_state=1)

forest_model.fit(train_X, train_y)

melb_preds = forest_model.predict(val_X)

print(mean_absolute_error(val_y, melb_preds))
```
